```python
import numpy as np

# 定义起始颜色和结束颜色
start_color = "#FFA500"  # 橙色
end_color = "#FFDAB9"  # 浅橙色

# 将起始颜色和结束颜色转换为RGB值
start_rgb = np.array([int(start_color[i:i+2], 16) for i in (1, 3, 5)]) / 255.0
end_rgb = np.array([int(end_color[i:i+2], 16) for i in (1, 3, 5)]) / 255.0

# 生成十个渐变颜色列表
colors = []
for i in range(10):
    t = i / 9.0  # 插值参数，范围从0到1
    rgb = (1 - t) * start_rgb + t * end_rgb  # 进行颜色插值计算
    color_hex = '#{:02x}{:02x}{:02x}'.format(
        int(rgb[0] * 255), int(rgb[1] * 255), int(rgb[2] * 255)
    )  # 将RGB值转换为16进制格式
    colors.append(color_hex)

# 打印颜色列表
print(colors)
```
