```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 定义年份和类别
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
categories = ['类别A', '类别B', '类别C', '类别D', '类别E', '类别F', '类别G']

# 生成模拟数据
data = []
for year in years:
    row = [year]
    total = 0
    for category in categories:
        value = np.random.randint(5, 15)
        row.append(value)
        total += value
    row = [value / total * 100 for value in row]
    data.append(row)

# 创建 DataFrame
df = pd.DataFrame(data, columns=['年份'] + categories)
df = df.set_index('年份')

# 绘制堆积条形图
plt.figure(figsize=(12, 8))
bottom = np.full(len(years),-0.2)  # 调整 bottom 参数

for category in categories:
    plt.barh(years, df[category], left=bottom, label=category, height=0.9)
    bottom += df[category]

plt.xlabel('percentage (%)')
plt.ylabel('年份')
plt.title('每年各类别百分比')

# plt.legend(ncol=5, frameon=False)

# 创建新的图例
# handles, labels = plt.gca().get_legend_handles_labels()
# plt.legend(handles, labels, loc='lower right')



# 自定义 x 轴刻度
xticks = list(range(0, 101, 4))
xtick_labels = [str(x) if x % 20 == 0 else '' for x in xticks]
plt.xticks(xticks, xtick_labels, minor=True)
plt.xticks(range(0, 101, 20), minor=False)

# # 自定义 y 轴刻度
# plt.yticks(ticks=[0, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000],labels=years)
plt.yticks(years)

plt.xlim(0, 100)
plt.ylim(2013.5,2023.5)
plt.show()
```
