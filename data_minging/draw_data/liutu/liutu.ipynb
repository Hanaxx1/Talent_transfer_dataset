```python
####时序的出入度

from collections import Counter
import pandas as pd
import json


country_name_tran_dict = {'United States':'the United States','Korea':'South Korea','the Congo':'Congo','DR Congo':'The Democratic Republic of Congo',
                          'Congo Brazzaville':'The Democratic Republic of Congo','St. Lucia':'Saint Lucia'}

def country_name_tran(country_name):
    if country_name in country_name_tran_dict:
        return country_name_tran_dict[country_name]
    else:
        return country_name

#时间戳转变 2000-2023
def convert_date_format(date_str):
    try:
        # 尝试将输入的字符串解析为日期
        date = pd.to_datetime(date_str, format='%Y-%m')
        # 确保日期在 2000 年到 2023 年之间
        if 2000 <= date.year <= 2023:
            # 将日期格式化为 'MM/YYYY' 形式
            formatted_date = date.strftime('%m/%Y')
            return formatted_date
        else:
            return None
    except ValueError:
        # 如果解析失败，则返回 None
        return None



origin_counter_time = {
    '01/2000': {'Others': 0},
}
destination_counter_time = {
    '01/2000': {'Others': 0},
}

country_name_tran_dict = {'United States':'the United States','Korea':'South Korea','the Congo':'Congo','DR Congo':'The Democratic Republic of Congo',
                          'Congo Brazzaville':'The Democratic Republic of Congo','St. Lucia':'Saint Lucia'}

country_shijieditu = []

path_file = '/home/qiuyang/workplace/swift/examples/pytorch/llm/my_data/data/c4_data_duiqi/c4_data.json'

with open(path_file, 'r') as input_file:
    # 逐行读取文件内容
    for line in input_file:
        # 解析JSON数据
        data = json.loads(line)

        origin = country_name_tran(data['origin'])
        destination = country_name_tran(data['destination'])

        if origin not in country_shijieditu:
            country_shijieditu.append(origin)
        if destination not in country_shijieditu:
            country_shijieditu.append(destination)

        timestamp = data['timestamp']
        timestamp = convert_date_format(timestamp)
        if timestamp!=None:
            if timestamp not in origin_counter_time:
                origin_counter_time[timestamp] = {}
            if origin not in origin_counter_time[timestamp]:
                origin_counter_time[timestamp][origin] = 1
            else:
                origin_counter_time[timestamp][origin] += 1

            if timestamp not in destination_counter_time:
                destination_counter_time[timestamp] = {}
            if destination not in destination_counter_time[timestamp]:
                destination_counter_time[timestamp][destination] = 1
            else:
                destination_counter_time[timestamp][destination] += 1
        # break



# print(country_shijieditu)
# print(len(country_shijieditu))


print(origin_counter_time.keys())
print(origin_counter_time)
print(destination_counter_time)
```

```python
##比率

file_name1 = '/home/qiuyang/workplace/world_map_drawing/Geographical heterogeneity/draw_data/liutu/month_year3.csv'
file_name2 = '/home/qiuyang/workplace/world_map_drawing/Geographical heterogeneity/draw_data/liutu/month_year4.csv'
file_name3 = '/home/qiuyang/workplace/world_map_drawing/Geographical heterogeneity/draw_data/liutu/month_year5.csv'

import pandas as pd
import numpy as np

# 读取第一个CSV文件
data1 = pd.read_csv(file_name1)

# 读取第二个CSV文件
data2 = pd.read_csv(file_name2)

# 提取第一列数据
first_column_data1 = data1.iloc[:, 0]
first_column_data2 = data2.iloc[:, 0]

# 从原始数据中移除第一列数据
data1 = data1.drop(data1.columns[0], axis=1)
data2 = data2.drop(data2.columns[0], axis=1)

# 对剩余数据进行相除，保留两位小数，并处理分母为0的情况
result = data1.div(data2, fill_value=0).round(2)

# 将无穷大（inf）值替换为0
result = result.replace([np.inf, -np.inf], 0)

# 将NaN值替换为0
result = result.fillna(0)

# 将第一列数据添加回结果中
result.insert(0, 'year', first_column_data1)
# 将处理后的结果保存到新的CSV文件中
result.to_csv(file_name3, index=False)
```
